```python
import spacy
import pandas as pd
```

```python
data = pd.read_csv(r'..\data\raw\Articulos_fechas.csv')
df = data.copy()
```

```python
nlp = spacy.load("es_core_news_lg")
numeros_letras = {
    "uno": 1, "dos": 2, "tres": 3, "cuatro": 4, "cinco": 5,
    "seis": 6, "siete": 7, "ocho": 8, "nueve": 9, "diez": 10,
    "once": 11, "doce": 12, "trece": 13, "catorce": 14, "quince": 15,
    "dieciséis": 16, "diecisiete": 17, "dieciocho": 18, "diecinueve": 19,
    "veinte": 20
}

def extraer_edad_recomendada(texto):
    if pd.isnull(texto):
        return "No especificado"

    texto = texto.lower()
    doc = nlp(texto)

    frases_contexto = ["a partir de", "para niños de", "recomendado para", "edad recomendada", "dirigido a", "ideal para", "niños desde"]

    for sent in doc.sents:
        sent_text = sent.text.lower()
        if any(frase in sent_text for frase in frases_contexto):
            for token in sent:
                if token.like_num:
                    try:
                        edad = int(token.text)
                    except ValueError:
                        edad = numeros_letras.get(token.text, None)

                    if edad is not None and 0 < edad <= 18:
                        return f"{edad} años"

    return "No especificado"
```

```python
df['edad'] = df['contenido'].apply(extraer_edad_recomendada)
```
